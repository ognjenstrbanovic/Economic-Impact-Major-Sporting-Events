```python
import requests
from pprint import pprint
from test import key
import numpy as np
import matplotlib.pyplot as plt
```

```python
GBP_rates= []
ZAR_rates = []
symbols = 'GBP, ZAR'
```

```python
before_date = '2008-08-12'
url = f'http://data.fixer.io/api/{before_date}?access_key={key}&symbols={symbols}'

response = (requests.get(url)).json()

ZAR_rate0 = response['rates']['ZAR']
ZAR_rates.append(ZAR_rate0)
```

```python
before_date = '2010-08-12'
url = f'http://data.fixer.io/api/{before_date}?access_key={key}&symbols={symbols}'

response = (requests.get(url)).json()

GBP_rate1 = response['rates']['GBP']
GBP_rates.append(GBP_rate1)
ZAR_rate1 = response['rates']['ZAR']
ZAR_rates.append(ZAR_rate1)
```

```python
during_date = '2012-08-12'
url2 = f'http://data.fixer.io/api/{during_date}?access_key={key}&symbols={symbols}'

response = (requests.get(url2)).json()

GBP_rate2 = response['rates']['GBP']
GBP_rates.append(GBP_rate2)
ZAR_rate2 = response['rates']['ZAR']
ZAR_rates.append(ZAR_rate2)
```

```python
after_date = '2014-08-12'
url3 = f'http://data.fixer.io/api/{after_date}?access_key={key}&symbols={symbols}'

response = (requests.get(url3)).json()

GBP_rate3 = response['rates']['GBP']
GBP_rates.append(GBP_rate3)
```

```python
print(GBP_rates)
print(ZAR_rates)
```

```python
x_axis = np.arange(len(GBP_rates))
x_axis
```

```python
plt.plot(x_axis, GBP_rates, color='b')
tick_locations = [value for value in x_axis]
labels = ['2010', '2012', '2014']
plt.xticks(tick_locations, labels, rotation='45')
plt.title('GBP vs USD')
plt.ylabel('Pound / $')
```

```python
plt.plot(x_axis, ZAR_rates, color='b')
tick_locations = [value for value in x_axis]
labels = ['2008', '2010', '2012']
plt.xticks(tick_locations, labels, rotation='45')
plt.title('Rand vs USD')
plt.ylabel('Rand / $')
```
